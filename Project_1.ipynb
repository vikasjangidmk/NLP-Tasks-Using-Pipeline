```python
import pandas as pd
import numpy as np
from transformers import pipeline
```

```python
cls = pipeline("sentiment-analysis")
```

```python
cls("I hate eat banana")
```

```python
cls("I love to be ai developer")
```

```python
clf = pipeline("zero-shot-classification")
```

```python
clf("this course all about transformers library",
    candidate_labels = ["education", "politics","bussiness",]
    )
```

```python
generator = pipeline("text-generation")
```

```python
generator("In this course, we will teach you how to")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "In this machine learning course, we will teach you how to",
    max_length=30,
    num_return_sequences=2,
)
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
ner("My name is Vikas Jangid form India and working as AI Engineer")
```
